# Analisis Sentimen menggunakan library NLTK
Setelah melewati tahap ekstraksi fitur, dataset kemudian akan di analisis sentimen dengan memberikan label Positif, Netral atau Negatif pada setiap tweet

## 1. Import Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'processed_tweets.csv'
dataset = pd.read_csv(file_path)

dataset.head()
```

## 2. Analisis Sentimen dengan Lexicon menggunakan Vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Inisialisasi analyzer
analyzer = SentimentIntensityAnalyzer()

# Fungsi untuk mendapatkan skor sentimen
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return 'positive'
    elif scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Terapkan analisis sentimen pada kolom 'full_text'
dataset['sentiment'] = dataset['full_text'].apply(get_sentiment)

# Lihat hasil
print(dataset[['full_text', 'sentiment']].head())
```

## 3. Simpan hasil analisis sentimen

```python
output_file_path = 'sentiment_analysis_results.csv'
dataset.to_csv(output_file_path, index=False)

print(f"Dataset with sentiment analysis saved to: {output_file_path}")
```
